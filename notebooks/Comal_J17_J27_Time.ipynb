# Load Comal River data and well data (contains streamflow, springflow, J-17 and J-27 max levels)

```python
df = pd.read_csv(os.path.join("..", "..", "data", "streamflow", "Springflow_All.csv"))
df.set_index(pd.to_datetime(df['datetime']), inplace=True)
df.dropna(inplace=True)
del df['datetime']
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,15))

x = df['MaxLevel_J17']
y = df['Comal Springflow']

ax.scatter(x, y)
ax.set_title('Comal Springflow vs. J-17 Max Level', size=20)
ax.set_ylabel("Comal Springflow (cfs)", size=20)
ax.set_xlabel("J-17 Max Level", size=20)
ax.grid()
plt.savefig(os.path.join("..", "..", "img_2019", "Comal_J17.png"))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,15))

x = df["Springflow"]
y = df["MaxLevel_J17"]

plt.scatter(x, y)
ax.set_title('Comal Springflow vs. J-27 Max Levels', size=20)
ax.set_xlabel("Comal Springflow (cfs)", size=20)
ax.set_ylabel("J-27 Max Level (feet above msl)", size=20)
ax.grid()
ax.legend()
plt.savefig(os.path.join("..", "..", "img_2019", "Comal_J27.png"))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))

x = df.index
y = df['Comal Springflow']

ax.plot(x, y)

ax.set_title('Comal Springflow vs. Time', size=20)
ax.set_ylabel("Comal Springflow (cfs)", size=20)
ax.set_xlabel("Date", size=20)
ax.grid()
ax.legend()
plt.savefig(os.path.join("..", "..", "img_2019", "Comal_Time.png"))
plt.show()
```
