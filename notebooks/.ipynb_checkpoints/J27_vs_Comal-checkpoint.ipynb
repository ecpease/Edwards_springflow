```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
j27 = pd.read_csv(os.path.join("..", "data", "dailydateRange_j27.csv"))
j27.set_index(pd.to_datetime(j27['datetime']), inplace=True)
j27.head()
```

```python
comal = pd.read_csv(os.path.join("..", "data", "USGS_streamgage", "main_ComalRvs.csv"))
comal.set_index(pd.to_datetime(comal['datetime']), inplace=True)
comal.shape
```

```python
df_main = comal.merge(j27, right_index=True, left_index=True)
df_main.to_csv(os.path.join("..", "data", "output_comalJ27.csv"))
df_main
```

```python
X, y = np.array(df_main['Comal River'].tolist()), np.array(df_main['MaxLevel'].tolist()) # make these list like arrays
```

```python
# X1 = []
# for i in range(len(X)):
#     X1.append(np.array([d[i], hour[i], month[i]])) # eventually we can add more input data here
# X1 = np.array(X1)
# print("X1 Shape is ", X1.shape)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.5, random_state=123)

train_X = train_X.reshape(-1, 1)
test_X  = test_X.reshape(-1, 1)
train_y = train_y.reshape(-1, 1)
test_y  = test_y.reshape(-1, 1)

print("Labels for training and testing data")
print("Train X shape: ", train_X.shape)
print("Train Y shape: ", train_y.shape)
print("Test  X shape: ", test_X.shape)
print("Test  Y shape: ", test_y.shape)
```

```python
k = 20 # ?? I dont know what is best to use yet
knn = KNeighborsRegressor(n_neighbors=k) # use regressor for this kind of data
knn.fit(train_X, train_y)
```

```python
future_j27 = [[500]] # Springflow at Comal Springs
future_j27 = np.array(future_j27)
```

```python
y_pred = knn.predict(future_j27)
print(y_pred)
```
