```python
upstream_guad = pd.read_csv(os.path.join("..", "data", "streamflow", "output", "USGS_drain_8165300.csv"))
upstream_guad = upstream_guad.set_index('datetime')
upstream_guad

```

```python
downstream_guad = pd.read_csv(os.path.join("..", "data", "streamflow", "output", "USGS_drain_8165500.csv"))
downstream_guad = downstream_guad.set_index("datetime")
downstream_guad
```

```python
df_guad = upstream_guad.merge(downstream_guad, left_index=True, right_index=True)
df_guad['site_no_upstream'] = df_guad['site_no_x']  
df_guad['site_no_downstream'] = df_guad['site_no_y']
df_guad['Q_up'] = df_guad['136364_00060_00003'] 
df_guad['Q_down'] = df_guad['136366_00060_00003'] 
df_guad = df_guad[['site_no_upstream', 'Q_up', 'site_no_downstream', 'Q_down']]

df_guad.head()
```

```python
df_guad['Q_diff'] = df_guad['Q_down'] - df_guad['Q_up']
df_guad
```

```python
fig, ax = plt.subplots()

x = df_guad.index
y_up = df_guad['Q_up']
y_down = df_guad['Q_down']
y_diff = df_guad['Q_diff']

plt.plot_date(x, y_diff)
plt.show()
```
