```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(os.path.join("..", "..", "data", "streamflow", "Springflow_All.csv"))
df['datetime'] = pd.to_datetime(df['datetime'])
df.head()
```

```python
df = df.resample('D', on='datetime').mean()
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15,7))

x = df.index
y = df["MaxLevel_J17"]

plt.plot(x, y)
ax.set_title('J-17 Max Water Levels 1986-2018', size=20)
ax.set_ylabel("J-17 Water Level (feet above msl)", size=20)
ax.set_xlabel("Year", size=20)
fig.tight_layout()
ax.grid()

plt.savefig(os.path.join("..", "..", "img_2019", "J17_time.png"))
plt.show()
```
